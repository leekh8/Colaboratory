```python
import numpy as np
```

```python
# @title 활성화 함수 (activation function) -> step function
def activation_function(x):
  if x <= 0:
    return 0
  else:
    return 1
```

```python
# @title 퍼셉트론 함수
def perceptron(x): # x는 외부에서 들어오는 데이터. 가중치가 10 절편이 -3
  w = 10
  b = -3
  y = w * x + b
  return activation_function(y)
```

```python
input_data = 10
a1 = perceptron(input_data)
a2 = perceptron(a1)
a2
```

```python
# @title AND 퍼셉트론 만들기
def AND_perceptron(x1, x2):
  w1 = 1
  w2 = 1
  b = -1
  y = w1 * x1 + w2 * x2 + b
  return activation_function(y)
```

```python
print(AND_perceptron(1, 1)) # 1
print(AND_perceptron(1, 0)) # 0
print(AND_perceptron(0, 1)) # 0
print(AND_perceptron(0, 0)) # 0
```

```python
# @title OR 퍼셉트론 만들기
def OR_perceptron(x1, x2):
  w1 = 1
  w2 = 1
  b = 0
  y = w1 * x1 + w2 * x2 + b
  return activation_function(y)
```

```python
print(OR_perceptron(1, 1)) # 1
print(OR_perceptron(1, 0)) # 1
print(OR_perceptron(0, 1)) # 1
print(OR_perceptron(0, 0)) # 0
```

```python
# @title NAND 퍼셉트론 만들기
def NAND_perceptron(x1, x2):
  w1 = -1
  w2 = -1
  b = 2
  y = w1 * x1 + w2 * x2 + b
  return activation_function(y)
```

```python
print(NAND_perceptron(1, 1)) # 0
print(NAND_perceptron(1, 0)) # 1
print(NAND_perceptron(0, 1)) # 1
print(NAND_perceptron(0, 0)) # 1
```

```python
# @title XOR을 위한 MLP 만들기 (다층 퍼셉트론)
def XOR_perceptron(x1, x2):
  return  AND_perceptron(NAND_perceptron(x1, x2),OR_perceptron(x1, x2))
```

```python
print(XOR_perceptron(1, 1)) # 0
print(XOR_perceptron(1, 0)) # 1
print(XOR_perceptron(0, 1)) # 1
print(XOR_perceptron(0, 0)) # 0
```

```python
# @title keras로 XOR MLP 만들기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense
```

```python
# @title 모델 생성
XOR_model = Sequential() # sequential은 일종의 뼈대
# 입력층 추가
# 입력될 데이터가 어떤 모양일지 미리 알려줌
XOR_model.add(InputLayer(input_shape = (2, )))
# 중간층 추가
# activation: 활성화 함수
XOR_model.add(Dense(units = 2, activation = 'sigmoid'))
# 출력층 추가
XOR_model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
XOR_model.summary()
```

```python
# @title 학습 데이터 생성
# @markdown # 미리 했어야 하지만 지금 함
X = np.array([[1, 1], [1, 0], [0, 1], [0, 0]])
y = np.array([0, 1, 1, 0])
```

```python
# @title 모델 학습
XOR_model.compile(loss = 'binary_crossentropy', optimizer = 'sgd', metrics = ['accuracy'])
XOR_model.fit(X, y, epochs = 3000)
```

```python
# @markdown # 너무 정확도 안나와서 optimazer 바꿈
XOR_model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
XOR_model.fit(X, y, epochs = 2000)
```

```python
# @title 출력층 가중치 확인
XOR_model.get_layer('dense_3').get_weights()
```
