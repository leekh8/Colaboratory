# Sigmoid
- 0-1 사이의 S 곡선의 함수
image.png

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_f(x):
  return 1 / (1 + np.exp(-x))

# sigmoid 함수 그리기
x = np.arange(-5.0 , 5.0, 0.1)
y = sigmoid_f(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

# Sigmoid 미분하기

```python
def sigmoid_d(x):
  return sigmoid_f(x) * (1 - sigmoid_f(x))

y = sigmoid_d(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.0)
plt.show()

y.max()
```

# tanh 함수

```python
y = np.tanh(x)

plt.plot(x, y)
plt.show()
```

# tanh 함수 미분하기

```python
def tanh_d(x):
  return (1-np.tanh(x))*(1+np.tanh(x))

y = tanh_d(x)

plt.plot(x, y)
plt.show()

y.max()
```

# ReLU

```python
def relu_f(x):
  return np.maximum(0, x)

y = relu_f(x)

plt.plot(x, y)
plt.show()

y.max()
```
